```python
data_series = pd.read_csv("data_day.csv", parse_dates =['Date'], index_col="Date")
```

```python
data_series.info()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing as HWES

hwes = HWES(data_series['Sales_Clean'])
fitted = hwes.fit()
print(fitted.summary())
```

```python
predict = fitted.predict(start = 0, end = 2783)
print(predict)
```

```python
MAPE = np.mean(np.abs((data_series['Sales_Clean']-predict)/data_series['Sales_Clean']))*100
print(MAPE, '%')
```

```python
import matplotlib as plt
import matplotlib.pyplot as plt

plt.figure(figsize=(19,7))
plt.plot(data_series['Sales_Clean'], "b", label = "actual")
plt.plot(predict, "r--", label = "Prediksi")
plt.legend()
```
